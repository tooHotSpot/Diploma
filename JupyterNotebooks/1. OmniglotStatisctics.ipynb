# Saving Omniglot background images as dictinary to the file

```python
def vanilla_load_dataset(folder):
    images_data = {}
    for alphabet in os.listdir(folder):
        images_data[alphabet] = {}
        alphabet_path = os.path.join(folder, alphabet)
        for character in os.listdir(alphabet_path):
            images_data[alphabet][character] = {}
            character_path = os.path.join(alphabet_path, character)
            for image in os.listdir(character_path):
                # Changed image_path to full_image_path
                full_image_path = os.path.join(character_path, image)
                images_data[alphabet][character][image] = cv2.imread(full_image_path, cv2.IMREAD_GRAYSCALE)
        print('Went through ', alphabet)
        
    print('Finished')
    return images_data
```

# Get knowledge about data in all_runs 

```python
def load_all_runs_images_dict():
    all_funs_folder = '../images/one-shot-classification/all_runs'
    file = '../Data/all_runs_images_dict.pkl'

    if os.path.isfile(file):
        return unpickle_it(file)
    
    images_dict = dict()
    for run_folder in os.listdir(all_funs_folder):
        for folder_path in ('training', 'test'):
            full_folder_path = os.path.join(all_funs_folder, run_folder, folder_path)
            for image_path in os.listdir(full_folder_path):
                full_image_path = os.path.normpath(os.path.join(full_folder_path, image_path))
                image = cv2.imread(full_image_path, cv2.IMREAD_GRAYSCALE)
                images_dict[full_image_path] = np.array(image, dtype=np.uint8)

    pickle_it(images_dict, file)
    return images_dict

def load_all_runs_annotation():
    all_funs_folder = '../images/one-shot-classification/all_runs'
    file = '../Data/all_runs_annotation.json'

    if os.path.isfile(file):
        return unjson_it(file)

    annotation_dict = dict()
    for run_folder in os.listdir(all_funs_folder):
        full_annotation_file_path = os.path.join(all_funs_folder, run_folder, 'class_labels.txt')
        with open(full_annotation_file_path) as f:
            lines = f.readlines()
        for line in lines:
            test, train = line.strip('\n').split(' ')
            annotation_dict[test] = train

    json_it(annotation_dict, file)
    return annotation_dict
```

```python
def brut_force_matching(images_dict, image_dict_name, all_runs_dict, recheck=False):

    intersected_alphabets = set()
    intersected_images = set()
    
    file = '../Data/eval_images_map.json'
    if os.path.exists(file):
        eval_images_map = unjson_it(file)
        print('Amount of keys in map: ', len(eval_images_map))
        
    if recheck:
        eval_images_map = dict()
        for k, key in enumerate(sorted(list(all_runs_dict))):
            key = os.path.normpath(key)
            for i, alphabet in enumerate(images_dict):
                for j, char in enumerate(images_dict[alphabet]):
                    for char_image_name in images_dict[alphabet][char]:
                        char_image = images_dict[alphabet][char][char_image_name]
                        char_image_name = os.path.normpath(char_image_name)

                        if np.linalg.norm(all_runs_dict[key] - char_image, ord='fro') == 0:
                            a = '/'.join(key.split('\\')[-4:])
                            b = os.path.join(image_dict_name, alphabet, char, char_image_name)
                            if a in eval_images_map.keys():
                                print('Intersection:', a,  'is ', eval_images_map[a], 'but ', b, ' is also.')
                            else:
                                eval_images_map[a] = b
                            
                            if k < 10:
                                print(a, ' = ', b)
        json_it(eval_images_map, file)

    
    for eval_image in eval_images_map:
        known_image = eval_images_map[eval_image]
        alphabet = known_image.split('/')[1] 
        intersected_alphabets.add(alphabet)
        if known_image in intersected_images:
            print('Image already chosen!')
        else:
            intersected_images.add(known_image)

    intersected_alphabets = sorted(intersected_alphabets)
    print('Amount of images chosen for the evaluation tasks: ', len(intersected_images))
    print('Some intersected images: ', list(intersected_images)[:2])
    print('Amount of alphabets chosen for the evaluation tasks: ', len(intersected_alphabets))
    
    for alphabet in intersected_alphabets:
        print(alphabet)
    return intersected_alphabets
    
```

# Checking method for fast loading of 400 evaluation one-shot learning trials (8000 verification tasks).

```python
def load_eval_trials(rewrite=False):
    """
    Method for fast loading of 400 evaluation one-shot learning trials, resulting in 8000 verification tasks.
    """ 
    
    all_funs_folder = '../images/one-shot-classification/all_runs'
    file = '../Data/all_runs_trials.pkl'
    
    if os.path.exists(file) and not rewrite:
        all_comparisons, all_comparisons_answers = unpickle_it(file)
    else:
        all_comparisons_list = []
        all_comparisons_answers_list = []

        for run_folder in os.listdir(all_funs_folder):
            annotation_file = os.path.join(all_funs_folder, run_folder, 'class_labels.txt')
            print('Here in ', run_folder, ' looking for the annotation file: ', os.path.normpath(annotation_file))
            with open(annotation_file) as f:
                lines = f.readlines()
            image_true_tuples_pairs = []
            for line in lines:
                test_image_path, train_image_path = line.strip().split()

                test_image_full_path = os.path.join(all_funs_folder, test_image_path)
                test_image = cv2.imread(test_image_full_path, cv2.IMREAD_GRAYSCALE)

                train_image_full_path = os.path.join(all_funs_folder, train_image_path)
                train_image = cv2.imread(train_image_full_path, cv2.IMREAD_GRAYSCALE)

                image_true_tuples_pairs.append([test_image, train_image])

            for i in range(20):
                for j in range(20):
                    all_comparisons_list.append([image_true_tuples_pairs[i][0], image_true_tuples_pairs[j][1]])
                    all_comparisons_answers_list.append(i == j)

        all_comparisons = np.expand_dims(np.array(all_comparisons_list, dtype=np.uint8), axis=-1)
        all_comparisons_answers = np.array(all_comparisons_answers_list, dtype=np.uint8).flatten()

        assert all_comparisons.shape == (8000, 2, 105, 105, 1), 'Incorrect all_comparisons images shape'
        assert all_comparisons_answers.shape == (8000,), 'Incorrect all_comparisons_answers shape'
        l = [all_comparisons, all_comparisons_answers]
        pickle_it(l, file)
    
    print('Loaded ', len(all_comparisons), ' image pairs with ', len(all_comparisons_answers), ' answers to them')
    return all_comparisons, all_comparisons_answers
```

```python
eval_trials, eval_trials_answers = load_eval_trials(rewrite=False)
```

```python
def random_check_eval_trials(eval_trials, eval_trials_answers):
    assert len(eval_trials) == len(eval_trials_answers) == 8000, 'Incorrect eval trials shape'
    chosen = np.random.choice(8000, 10)
    fig, ax = plt.subplots(10, 2, figsize=(5, 15), sharex=True)

    for i, pair_num in enumerate(chosen):
        image_pair = eval_trials[pair_num]
        test_image = image_pair[0].reshape(105, 105)
        train_image = image_pair[1].reshape(105, 105)
        pair_type = eval_trials_answers[pair_num]

        ax[i, 0].set_xlabel('                                                    ' + 
                            ('Same chars' if pair_type == 1 else 'Different chars'))        
        ax[i, 0].imshow(test_image, cmap='gray')
        ax[i, 0].set_xticks([])
        ax[i, 0].set_yticks([])
        
        ax[i, 1].imshow(train_image, cmap='gray')
        ax[i, 1].set_xticks([])
        ax[i, 1].set_yticks([])
                
    plt.tight_layout()
    plt.show()
```
